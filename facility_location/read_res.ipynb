```python
import matplotlib.pyplot as plt

rounding = "seq"

def read_text(input_file):
    with open(input_file, "r") as f:
        lines = f.readlines()
    train_loss_list = []
    test_obj_seq_list = []
    test_obj_grd_list = []
    for line in lines:
        _, train_loss, test_obj_seq, test_obj_grd = line.split()
        train_loss_list.append(float(train_loss))
        test_obj_seq_list.append(float(test_obj_seq))
        test_obj_grd_list.append(float(test_obj_grd))
    return train_loss_list, test_obj_seq_list, test_obj_grd_list

increase_ratio = 1.5

# Increase all font sizes for PDF readability
plt.rcParams.update({
    'font.size':        14 * increase_ratio,   # base font size
    'axes.titlesize':   16 * increase_ratio,
    'axes.labelsize':   14 * increase_ratio,
    'xtick.labelsize':  12 * increase_ratio,
    'ytick.labelsize':  12 * increase_ratio,
    'legend.fontsize':  12 * increase_ratio,
})

labels = ["Orig", "=10.0", "=1.0", "=0.1", "=0.01", "=0.001"]

settings2results = dict()
settings2results["Orig"] = read_text("logs/fl_orig_100/lr1e-1_reg1e-1_orig.txt")
settings2results["=10.0"] = read_text(f"logs/fl_{rounding}_100/lr1e-1_reg1e-1_temp10.0.txt")
settings2results["=1.0"] = read_text(f"logs/fl_{rounding}_100/lr1e-1_reg1e-1_temp1.0.txt")
settings2results["=0.1"] = read_text(f"logs/fl_{rounding}_100/lr1e-1_reg1e-1_temp1e-1.txt")
settings2results["=0.01"] = read_text(f"logs/fl_{rounding}_100/lr1e-1_reg1e-1_temp1e-2.txt") 
settings2results["=0.001"] = read_text(f"logs/fl_{rounding}_100/lr1e-1_reg1e-1_temp1e-3.txt")

# Define line styles and thicknesses for better accessibility
line_styles = ['-', '--', ':', '-.', (0, (3, 1, 1, 1)), (0, (5, 1))]
line_width = 2.5  # Thicker lines

# Create a figure with two subplots side-by-side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 5))

# Plot training loss in the first subplot
for i, label in enumerate(labels):
    train_loss_list, test_obj_seq_list, test_obj_grd_list = settings2results[label]
    ax1.plot(train_loss_list, label=label, linestyle=line_styles[i], linewidth=line_width)
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Training Loss")
ax1.grid(True, linestyle="--", alpha=0.5)

# Plot test objective in the second subplot
for i, label in enumerate(labels):
    train_loss_list, test_obj_seq_list, test_obj_grd_list = settings2results[label]
    ax2.plot(test_obj_seq_list, label=label, linestyle=line_styles[i], linewidth=line_width)
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Test Objective")
ax2.grid(True, linestyle="--", alpha=0.5)

# Create a single legend at the top of the figure with reduced space between items
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 1.08), 
           ncol=6, columnspacing=0.8, handletextpad=0.4)

plt.tight_layout()
# plt.subplots_adjust(top=0.85)  # Add space at the top for the common legend
plt.savefig("fl_seq_100.pdf", bbox_inches='tight', dpi=300)
# plt.show()
```
